```python
import numpy as np

xx = np.zeros([2,3], np.complex)
xx[0,0] = 10.2 + 4.0 * 1j
xx[0,1] = 1.2 + 2.0 * 1j
xx[0,2] = 3.2 + 4.0 * 1j
xx[1,0] = 4.2 + 3.0 * 1j
xx[1,1] = 1.2 + 5.0 * 1j
xx[1,2] = 8.2 + 4.0 * 1j
print(xx)

comp_datatype = np.dtype([
            ('real', np.float),
            ('imaginary', np.float)
        ])

yy = np.zeros([2,3], dtype=comp_datatype)
for id_el in range(2):
    for id_e2 in range(3):
        yy[id_el, id_e2] = (xx[id_el, id_e2].real, xx[id_el, id_e2].imag)
print(yy)
```
